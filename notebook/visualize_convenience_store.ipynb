```python
import os.path

import dill
import overpy
```

```python
api = overpy.Overpass()
```

```python
prefs = ''
with open('pref.txt') as f:
    for row in f.readlines():
        prefs += '{}|'.format(row.rstrip())
prefs = prefs[:-1]
print(prefs)
```

```python
query = """
area["name"~"{}"];
node(area)["shop"="convenience"];
out;
""".format(prefs)
print(query)

if not os.path.exists('overpy_result.pkl'):
    with open('overpy_result.pkl', 'wb') as f:
        result = api.query(query)
        dill.dump(result, f)
```

```python
with open('overpy_result.pkl', 'rb') as f:
    result = dill.load(f)
```

```python
len(result.nodes)
```

```python
coord_list = []
for node in result.nodes:
#     print(node.tags)
    coord_list.append([node.lon, node.lat])
```

```python
from bokeh.plotting import figure
from bokeh.io import show, output_file, output_notebook, push_notebook

lon = []
lat = []
for coord in coord_list:
    lon.append(coord[0])
    lat.append(coord[1])

plot = figure()
plot.grid.visible = False
plot.axis.visible = False
plot.circle(lon, lat, color='#4c4c4c', size=0.5)

output_notebook()
handle = show(plot, notebook_handle=True)
push_notebook(handle=handle)
```
